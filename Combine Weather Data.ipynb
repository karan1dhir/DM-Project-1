```python
import streamlit as st
import folium
import pandas as pd
import numpy as np
import os
```

```python
flag = True
i = 1

# https://stackoverflow.com/questions/10377998/how-can-i-iterate-over-files-in-a-given-directory#10378012

df_details = pd.read_csv(os.path.join(os.getcwd(), "StormEvents_details-ftp_v1.0_d1950_c20170120.csv"))
df_fatalities = pd.read_csv(os.path.join(os.getcwd(), "StormEvents_fatalities-ftp_v1.0_d1971_c20160223.csv"))
df_locations = pd.read_csv(os.path.join(os.getcwd(), "StormEvents_locations-ftp_v1.0_d2020_c20201216.csv"))

for filename in os.listdir(os.path.join(os.getcwd(), "CSV Files")):
    filepath = os.path.join(os.getcwd(), "CSV Files", filename)

    print("Appending File ", i, ": ", filename)
    i += 1

    if "details" in filepath:

        df_temp = pd.read_csv(filepath)
        df_details = df_details.append(df_temp)

        
for filename in os.listdir(os.path.join(os.getcwd(), "CSV Files")):
    filepath = os.path.join(os.getcwd(), "CSV Files", filename)

    print("Appending File ", i, ": ", filename)
    i += 1
    
    if "fatalities" in filepath:  

        df_temp = pd.read_csv(filepath)
        df_fatalities = df_fatalities.append(df_temp)

        
for filename in os.listdir(os.path.join(os.getcwd(), "CSV Files")):
    filepath = os.path.join(os.getcwd(), "CSV Files", filename)

    print("Appending File ", i, ": ", filename)
    i += 1
    
    if "locations" in filepath:     

        df_temp = pd.read_csv(filepath)
        df_locations = df_locations.append(df_temp)    

```

```python
df = df_details.merge(df_locations, on = ["EVENT_ID"], how = "left") 
df = df.merge(df_fatalities, on = ["EVENT_ID"], how = "left") 
```

```python
df.to_csv("compiled_data.csv", compression='gzip')
```
