```python
import numpy as np
import mnist
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, MaxPooling3D, Conv3D, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import image_dataset_from_directory
from keras.callbacks import ModelCheckpoint
import tensorflow as tf
```

```python
### Derived from https://victorzhou.com/blog/keras-cnn-tutorial/





### With help from https://machinelearningmastery.com/how-to-load-large-datasets-from-directories-for-deep-learning-with-keras/
datagen = ImageDataGenerator(
                        featurewise_center=False,
                        featurewise_std_normalization=False,
                        rotation_range=10,
                        width_shift_range=0.1,
                        height_shift_range=0.1,
                        zoom_range=.1,
                        horizontal_flip=True,
                        validation_split=.3)

train_it = datagen.flow_from_directory('/kaggle/input/yawn-eye-dataset-new/dataset_new/train', batch_size=32, color_mode = "grayscale")

print(train_it)

num_filters = 8
filter_size = 16
pool_size = 3

### https://stackoverflow.com/questions/45393429/keras-how-to-save-model-and-continue-training

filepath = "/kaggle/working/model.h5"
checkpoint = ModelCheckpoint(filepath, monitor='accuracy', verbose=1, save_best_only=False, mode='max')
callbacks_list = [checkpoint]


    # Build the model.
model = Sequential([
      Conv2D(num_filters, filter_size, input_shape=(256, 256, 1)),
      MaxPooling2D(pool_size=pool_size),
      Conv2D(16, 8),
      MaxPooling2D(pool_size=pool_size),
      Conv2D(64, 4),
      MaxPooling2D(pool_size=pool_size),
      BatchNormalization(),
      Flatten(),
      Dense(600),
      Dropout(.1),
      Dense(200),
      Dropout(.1),
      Dense(200),
      Dropout(.1),
      Dense(4, activation='softmax'),
])

    # Compile the model.
model.compile(
      'adam',
      loss='categorical_crossentropy',
      metrics=['accuracy'],
)

# Train the model.

model.fit(train_it, steps_per_epoch=8, epochs = 10000, callbacks=callbacks_list)
```
